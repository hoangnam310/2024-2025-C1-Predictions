```python
df_matches_result = pd.read_html("https://fbref.com/en/comps/8/schedule/Champions-League-Scores-and-Fixtures",
                                 attrs={"id":"sched_all"})[0]
```

```python
df_matches_result
```

```python
df_matches_result.head()
```

```python
df_result_clean = df_matches_result[df_matches_result['Score'].notna()].copy()
```

```python
df_result_clean[df_result_clean['Score'].str.lower() == 'score']
```

```python
df_result_clean.head()
```

```python
df_result_clean[['HomeGoals', 'AwayGoals']] = df_result_clean['Score'].str.split('–', expand=True)
```

```python
df_result_clean
```

```python
df_result_clean['HomeGoals'].unique()
```

```python
df_result_clean = df_result_clean[df_result_clean['HomeGoals'] != 'Score']
df_result_clean = df_result_clean[df_result_clean['AwayGoals'] != 'Score']
```

```python
df_result_clean['HomeGoals'].unique()
```

```python
df_result_clean['HomeGoals'] = df_result_clean['HomeGoals'].str.extract(r'(?:\(\d+\)\s*)?(\d+)$')
```

```python
df_result_clean['AwayGoals'] = df_result_clean['AwayGoals'].str.extract(r'(?:\(\d+\)\s*)?(\d+)$')
```

```python
df_result_clean['HomeGoals'].unique()
```

```python
df_result_clean['AwayGoals'].unique()
```

```python
df_result_clean = df_result_clean[df_result_clean['AwayGoals'] != 'NaN']
```

```python
nan_rows = df_result_clean[df_result_clean['AwayGoals'].isna()]
nan_rows
```

```python
df_result_clean.at[178, 'AwayGoals'] = 1
df_result_clean.at[184, 'AwayGoals'] = 0
```

```python
df_result_clean['HomeGoals'] = df_result_clean['HomeGoals'].astype(int)
df_result_clean['AwayGoals'] = df_result_clean['AwayGoals'].astype(int)
```

```python
type_series = df_result_clean['HomeGoals'].apply(lambda x: type(x).__name__)
type_series_2 = df_result_clean['AwayGoals'].apply(lambda x: type(x).__name__)
type_series.value_counts()
type_series_2.value_counts()
```
